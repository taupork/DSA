```python
import re
import os
import pandas as pd

# HgB_8.9gdl_Individual02_02.jpeg


def synthetic_id_generator(start=100):
  while True:
    yield start
    start += 1

id_gen = synthetic_id_generator()

def parse_filename(filename, id_gen):
    name, ext = os.path.splitext(filename)
    parts = name.split("_")


    hb_value = None
    ethnicity = None
    individual_id = 10
    image_no = 0

    hb_match = re.search(r'(\d+(\.\d+)?)gdl', filename)
    if (hb_match):
      hb_value = float(hb_match.group(1))

    ethnicities = ["ChineseOrigin", "MiddleEasternOrigin"]
    for e in ethnicities:
      if e.lower() in filename.lower():
        ethnicity = e
        break

    indiv_match = re.search(r'Individual(\d+)', filename)
    if (indiv_match):
      individual_id = int(indiv_match.group(1))
    else:
      individual_id = next(id_gen)

    img_match = re.search(r'\d+_(\d+)\.', filename)
    if (img_match):
      image_no = int(img_match.group(1))

    return {
        "Filename" : filename,
        "Hb" : hb_value,
        "Ethnicity" : ethnicity,
        "IndividualID" : individual_id,
        "Image No" : image_no,
        "Extension" : ext.lower().strip(".")
    }
```

```python
# parse_filename("HgB_8.9gdl_Individual02_02.jpeg")
```

```python
from pathlib import Path
def extract_data(folder):
  files = [f for f in Path(folder).iterdir() if f.is_file()]

  parsed = [parse_filename(str(f), id_gen) for f in files]

  return pd.DataFrame.from_records(parsed)


```
