```python
import pandas as pd

CSV_PATH = "/content/drive/MyDrive/DSA_Comp/dataset.csv"
df = pd.read_csv(CSV_PATH)
df
```

```python
import os, torch, torch.nn as nn, numpy as np, pandas as pd, xgboost as xgb, optuna
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- Dataset Class ---
class HbImageDataset(Dataset):
    def __init__(self, df, transform=None, path_col="Filename"):
        self.df, self.transform, self.path_col = df, transform, path_col
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        img = Image.open(self.df.iloc[idx][self.path_col]).convert("RGB")
        if self.transform: img = self.transform(img)
        return img, self.df.iloc[idx]["Hb"]

# --- Transforms ---
train_transform = transforms.Compose([
    transforms.Resize((224,224)), transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15), transforms.ColorJitter(0.2,0.2),
    transforms.ToTensor(), transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])
val_transform = transforms.Compose([
    transforms.Resize((224,224)), transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

# --- Feature Extraction ---
def extract_features(df, transform, batch_size=4, path_col="Filename"):
    dataset, loader = HbImageDataset(df, transform, path_col), DataLoader(
        HbImageDataset(df, transform, path_col),
        batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True
    )
    model = nn.Sequential(*list(models.resnet50(pretrained=True).children())[:-1]).eval().to(device)
    feats, targets = [], []
    with torch.no_grad():
        for imgs, hb in loader:
            imgs = imgs.to(device, non_blocking=True)
            out = model(imgs).view(imgs.size(0), -1)
            feats.append(out.cpu().numpy())
            targets.extend(hb.numpy())
    return np.vstack(feats), np.array(targets)

# --- Metadata concatenation ---
def combine_metadata(features, df):
    if "IndividualID" in df.columns and "ImageNo" in df.columns:
        return np.hstack([features, df[["IndividualID","ImageNo"]].values])
    return features

# --- Train/Val Split ---
df_train, df_val = train_test_split(df, test_size=0.2, random_state=42)

# --- Extract Features ---
X_train, y_train = extract_features(df_train, train_transform)
X_train = combine_metadata(X_train, df_train)
X_val, y_val = extract_features(df_val, val_transform)
X_val = combine_metadata(X_val, df_val)

# --- XGBoost + Optuna ---
def objective(trial, X, y):
    params = {
        "n_estimators": trial.suggest_int("n_estimators",50,500),
        "max_depth": trial.suggest_int("max_depth",2,10),
        "learning_rate": trial.suggest_float("learning_rate",0.01,0.3,log=True),
        "subsample": trial.suggest_float("subsample",0.5,1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree",0.5,1.0),
        "random_state": 42
    }
    rmses = []
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    for train_idx, val_idx in kf.split(X):
        model = xgb.XGBRegressor(**params)
        model.fit(X[train_idx], y[train_idx])
        y_pred = model.predict(X[val_idx])
        rmses.append(np.sqrt(mean_squared_error(y[val_idx], y_pred)))
    return np.mean(rmses)

study = optuna.create_study(direction="minimize")
study.optimize(lambda t: objective(t, X_train, y_train), n_trials=30)
best_params = study.best_params

# --- Train Final Model ---
final_model = xgb.XGBRegressor(**best_params)
final_model.fit(X_train, y_train)
y_val_pred = final_model.predict(X_val)
rmse_val = np.sqrt(mean_squared_error(y_val, y_val_pred))
print("Validation RMSE:", rmse_val)

# --- Plot Predictions ---
plt.scatter(y_val, y_val_pred)
plt.plot([min(y_val), max(y_val)], [min(y_val), max(y_val)], 'r--')
plt.xlabel("Actual Hb"); plt.ylabel("Predicted Hb")
plt.title("XGBoost Predictions vs Actual Hb"); plt.show()
```
